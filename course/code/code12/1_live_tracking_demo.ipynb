# start in virtual python environment: 
 mlflow server --host 127.0.0.1 --port 5000

# Demo code to be tracked
Now run this, which is just generating output

```python
import time
import math
import mlflow
from mlflow.tracking import MlflowClient

mlflow.set_tracking_uri("http://localhost:5000")
mlflow.set_experiment("Live Sine Demo")

# ------------------------------------------------------------
# Persistent run_id handling
# ------------------------------------------------------------
if "RUN_ID" not in globals():
    run = mlflow.start_run(run_name="sine times cosine")
    RUN_ID = run.info.run_id
    print("Created new run:", RUN_ID)
else:
    run = mlflow.start_run(run_id=RUN_ID)
    print("Reusing run:", RUN_ID)

# ------------------------------------------------------------
# Determine last logged step
# ------------------------------------------------------------
client = MlflowClient()
history = client.get_metric_history(RUN_ID, "sine")

if history:
    STEP_OFFSET = max(m.step for m in history) + 1
else:
    STEP_OFFSET = 0

print("Starting at step:", STEP_OFFSET)

# ------------------------------------------------------------
# Log new data
# ------------------------------------------------------------
start_time = time.time()

for i in range(60):
    t = time.time() - start_time
    step = STEP_OFFSET + i
    mlflow.log_metric("sine", math.sin(t)*math.cos(t/10), step=step)
    time.sleep(1)

mlflow.end_run()
```
