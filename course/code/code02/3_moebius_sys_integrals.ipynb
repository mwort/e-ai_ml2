# 1. Some sophisticated plot

```python
import numpy as np, matplotlib.pyplot as plt
from matplotlib.colors import LightSource

u = np.linspace(0, 2*np.pi, 320); v = np.linspace(-0.6, 0.6, 80); U, V = np.meshgrid(u, v)

# Mbius-like twisted surface with extra complexity
X = (1 + V*np.cos(U/2)) * np.cos(U); Y = (1 + V*np.cos(U/2)) * np.sin(U)
Z = V*np.sin(U/2) + 0.3*np.sin(3*U)

ls = LightSource(azdeg=22, altdeg=65); rgb = ls.shade(Z, cmap=plt.cm.viridis, vert_exag=0.9)
plt.figure(figsize=(12,12)); ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, facecolors=rgb, linewidth=0); ax.set_axis_off(); plt.show()
```

# 3. Some Mathematics

We define a function by integration:

$$
f(x) := \int_{0}^{x} g(y)\,dy
$$

In practice, we often **do not know \( g \) analytically** and instead:
- sample it on a grid
- integrate numerically
- visualize the result

```python
import numpy as np
import matplotlib.pyplot as plt

# Define g(y)
def g(y):
    return np.sin(y) * np.exp(-0.2*y)

# Discretize domain
y = np.linspace(0, 10, 500)
g_vals = g(y)

# Numerical integration (cumulative integral)
f_vals = np.cumsum(g_vals) * (y[1] - y[0])

# Plot g and f
plt.figure(figsize=(8, 4))
plt.plot(y, g_vals, label=r"$g(y)$")
plt.plot(y, f_vals, label=r"$f(x)=\int_0^x g(y)\,dy$")
plt.xlabel("x")
plt.legend()
plt.grid()
plt.show()
```
